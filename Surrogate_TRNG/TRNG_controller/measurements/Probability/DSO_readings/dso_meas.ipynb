```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import butter, filtfilt
from scipy.optimize import curve_fit
import csv
```

```python
time1 = []
driv_sig = []
with open(f"Pplot_C2_sd1.2.csv", "r", encoding="utf-8") as file:
        csv_reader = csv.reader(file)
        for i, row in enumerate(csv_reader):
            if i >= 12:
                time1.append(float(row[0]))
                driv_sig.append(float(row[1]))
```

```python
time2 = []
mod_sig = []
with open(f"Pplot_C1_sd1.2.csv", "r", encoding="utf-8") as file: 
        csv_reader = csv.reader(file)
        for i, row in enumerate(csv_reader):
            if i >= 12:
                time2.append(float(row[0]))
                mod_sig.append(float(row[1]))
```

```python
plt.figure(figsize=(20,5))
plt.plot(time1,driv_sig)
plt.plot(time2,mod_sig)
plt.savefig("Pplot_hardware_meas.pdf")
```
