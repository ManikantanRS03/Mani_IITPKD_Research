```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
alpha = -0.15e4
beta = 0.75e12
#how?? alpha and beta

R = 1e4
RL = 0
rho = 1e-6
#how?? rho
#how the physical parameters of the core is taken into account in the calculation
```

```python
n = 50
Ival = 3
#I = []
#I.extend(Ival*np.ones(50))
#I.extend(-Ival*np.ones(50))
#I.extend(Ival*np.ones(50))
# step current with 3A and -3A for 150 steps

# triangluar wave
I = []
t1 = np.linspace(0, 1, 500)
I = 3*signal.sawtooth(2*np.pi*6*t1,0.5)
print(type(I))
plt.plot(t1,I)
plt.show

VL = [0]*len(I)
IL = [0]*len(I)
V = [0]*len(I)
ILguess=[0,0]
phi = [0]*len(I)
L = [0]*len(I)
```

```python
#t = np.arange(0,dt*len(I),dt) # 0 to 1.5us timescale in 10ns steps
#t = np.linspace(0, dt*len(I), len(I))
t = []
dt = 1e-8 #??? why
for i in range(len(I)):
    t.append(i*dt)
print(t)
```

```python
dt = 1e-9
Th = 1e-7
IL[0] = 0
phi[0] = 0
```

```python
for i in range(1,len(I),1):
    ILguess[0] = 0.5*I[i]
    diff = 1
    while(diff>Th):
        print("diff = ", diff)
        VL[i] = R*(I[i]-ILguess[0])-ILguess[0]*RL
        V[i] = R*(I[i] - ILguess[0])
        phi[i]=phi[i-1]+VL[i]*dt
        print("ILguess = ", ILguess[0])
        ILguess[1]=(phi[i]-phi[i-1])*(rho/dt) + 2*alpha*phi[i-1]+4*beta*phi[i-1]**3
        diff=abs(ILguess[1]-ILguess[0])
        ILguess[0] = ILguess[1]
    IL[i]=ILguess[1]
    print(IL[i])
    L[i]=(phi[i]-phi[i-1])/(IL[i]-IL[i-1])
    
```

```python
f = plt.figure(1)
fig, axes = plt.subplots(5,1)

axes[0].plot(t,I)
axes[0].set_xlabel('t')
axes[0].set_ylabel('I')

axes[1].plot(t,V)
axes[1].set_xlabel('t')
axes[1].set_ylabel('V')

axes[2].plot(t,VL)
axes[2].set_xlabel('t')
axes[2].set_ylabel('VL')

axes[3].plot(t,IL)
axes[3].set_xlabel('t')
axes[3].set_ylabel('IL')

axes[4].plot(t,phi)
axes[4].set_xlabel('t')
axes[4].set_ylabel('phi')

#axes[5].plot(t,L)
#axes[5].set_xlabel('t')
#axes[5].set_ylabel('L')

plt.show()
```
