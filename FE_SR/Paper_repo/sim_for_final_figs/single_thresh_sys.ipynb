```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python
def threshold_sys(sig, th):
    out = np.zeros(len(sig))
    for i in range(len(sig)):
        if(sig[i] > th):
             out[i] = sig[i] - th
    return out
```

```python
t = np.linspace(0, 0.07, 10000)

f_sig = 100
A_sig = 1
sd = 0.75

sig = A_sig*np.sin(2*np.pi*f_sig*t)
noise = np.random.normal(0, sd, len(t))

V_th = 2
sig_in = sig+noise
out_sig = threshold_sys(sig_in, 2)

plt.figure(figsize=(15,10))
plt.subplot(2,2,1); plt.plot(t, sig_in, c = "blue");  plt.plot(t, sig, c = "red"); plt.axhline(y=V_th, linestyle = "--", c = "black"); plt.ylim([-4,4])
plt.subplot(2,2,2); plt.plot(t, out_sig,  c = "black"); plt.ylim([0,3])
plt.subplot(2,2,3); plt.plot(t, sig, c = "green"); plt.ylim([-4,4])
plt.subplot(2,2,4); plt.plot(t, noise, c = "green"); plt.ylim([-4,4])
plt.savefig("single_threshold.pdf")
```

```python
fs = 1/(t[1]-t[0])
freq, psd = sc.signal.periodogram(out_sig, fs)
```

```python
plt.plot(freq, psd); plt.yscale("log"); plt.xscale("log")
```
