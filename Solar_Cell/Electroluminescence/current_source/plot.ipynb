```python
import matplotlib.pyplot as plt
from ltspice import Ltspice

# Specify the path to the .raw file
raw_file_path = "current_source.raw"

# Load the LTspice raw file
lt = Ltspice(raw_file_path)
lt.parse()  # Parse the raw file

# Extract time, voltage signal, and current
time = lt.get_time()
signal = lt.get_data("V(n006)")  # Replace "V(n006)" with the desired voltage signal name
current = lt.get_data("I(D1)")   # Replace "I(D1)" with the desired current signal name

# Plot the waveform with two y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the voltage signal on the primary y-axis
ax1.plot(time, signal, label="V(n006)", color="blue")
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Voltage (V)", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")
ax1.grid(True)

# Create a secondary y-axis for the current
ax2 = ax1.twinx()
ax2.plot(time, current,'--', label="I(D1)", color="red")
ax2.set_ylabel("Current (A)", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Add title and legends
fig.suptitle("LTspice Simulation Waveform")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# Show the plot
plt.savefig('current_waveform.pdf')
plt.show()
```
