```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Parameters from datasheet
P_total = 250  # from data sheet (mW)
I_radiant = 83  # from data sheet (mw/sr)
fwhm = 34  # from data sheet (nm)
lambda_peak = 525  # from data sheet (nm)
sigma = fwhm / (2 * np.sqrt(2 * np.log(2)))

measured_intensity = 0.496  # measured (mW/cm^2)
measured_distance = 10  # cm

# Gaussian function definition
def gaussian(lambda_val, Pmax, lambda_peak, sigma):
    return Pmax * np.exp(- (lambda_val - lambda_peak) ** 2 / (2 * sigma ** 2))


# Calculated intensity data
R_values = [2, 5, 10, 15, 30]  # cm
#R_values = [2,5,10] 
C_measured = measured_intensity * measured_distance ** 2  # Finding the value of C by fitting the measured data (mW)
C_R2_values = [C_measured / R ** 2 for R in R_values]  # I = C/R^2 (mW/cm^2)

lambda_vals = np.linspace(450, 600, 500)

plt.figure(figsize=[15, 7.5])

# Plotting the calculated Gaussian intensity profiles
for i, R in enumerate(R_values):
    P0 = C_R2_values[i]
    Pmax = P0 / (((math.sqrt(2 * math.pi)) * sigma)) * 1000
    intensity_profile = gaussian(lambda_vals, Pmax, lambda_peak, sigma)
    I = np.trapz(intensity_profile, lambda_vals)
    plt.subplot(1, 2, 1)
    plt.plot(lambda_vals, intensity_profile, linestyle='-', label=f'Calculated: d = {R} cm, I ={I:.2f} $\mu W/cm^2$')
    print(f'Calculated: d = {R} cm, I ={I:.2f} $\mu W/cm^2$')
    plt.title("Intensity Profile (Calculated vs Measured)")
    plt.xlabel("Wavelength (nm)")
    plt.ylabel("Intensity ($\mu W/cm^2/nm$)")

    plt.subplot(1, 2, 2)
    plt.scatter(R, P0, color='blue', marker='o', label='Calculated' if i == 0 else "")
    plt.title("LED Intensity at Different Distances")
    plt.xlabel("Distance (cm)")
    plt.ylabel("Intensity ($\mu W/cm^2$)")

# Plotting the measured data
for i, R in enumerate(R_values):
    # Reading the data from the file
    filename = f"data\\520nm_300mA_{R}cm_1LED.txt"
    wavelen = []
    ints_nm = []

    with open(filename, 'r') as file:
        for line in file:
            parts = line.split()
            wavelen.append(float(parts[0]))
            ints_nm.append(float(parts[1]))

    I_measured = np.trapz(ints_nm, wavelen)
    fro = 500  # Adjusted index to use full range (ensure data is accurate in file)
    to = 900

    # Plotting the measured data with different line style
    plt.subplot(1, 2, 1)
    plt.plot(wavelen[fro:to], ints_nm[fro:to], linestyle='--', label=f'Measured: d = {R} cm, I = {I_measured:.2f} uW/cm²')
    print(f'Measured: d = {R} cm, I = {I_measured:.2f} uW/cm²')
    plt.grid()
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.scatter(R, I_measured * 1e-3, color='orange', marker='x', label='Measured' if i == 0 else "")

plt.legend()
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.savefig("fig2.pdf")
plt.show()

```
